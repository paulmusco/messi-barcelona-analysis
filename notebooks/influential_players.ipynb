# Starting XIs

Which players has Messi played with the most? Which are the most influential in terms of his output? Which are the most influential in terms of Barcelona results?

```python
events_cleaned = pd.read_csv('../data/cleaned-data/events_cleaned.csv', low_memory=False)
```

```python
events_cleaned.head()
```

Subset 'events_cleaned' dataframe to show only Barcelona starting XI line-ups

```python
starting_xis = events_cleaned[(events_cleaned['team_name'] == 'Barcelona') & (events_cleaned['type_id'] == 35)]
```

```python
starting_xis.head()
```

```python
len(starting_xis)
```

Look into null values

```python
starting_xis.isnull().sum()
```

Drop columns that contain only null values

```python
null_cols = starting_xis.isnull().sum()
drop_cols = list(null_cols[null_cols == 485].index)
starting_xis = starting_xis.drop(drop_cols, axis=1)
starting_xis.head()
```

Subset 'starting_xis' dataframe to show only the 'player_' columns

```python
players = starting_xis[starting_xis.columns[-11:]].copy()
```

```python
players.head()
```

Subset 'starting_xis' dataframe to show only the 'match info' columns

```python
match_info = starting_xis.iloc[:,:-11].copy()
```

```python
match_info.head()
```

```python
players.iloc[0, 0]
```

```python
type(players.iloc[0, 0])
```

Convert string to dictionary

```python
import ast
for col in players.columns:
    players[col] = [ast.literal_eval(i) for i in players[col]]
```

```python
type(players.iloc[0, 0])
```

Select only the player name in each field

```python
for col in players.columns:
    players[col] = [players[col].iloc[i]['player'] for i in range(len(players[col]))]
```

```python
players.head()
```

```python
for col in players.columns:
    players[col] = [players[col].iloc[i]['name'] for i in range(len(players[col]))]
```

```python
players.head()
```

Concat the 'match_info' and 'players' dataframes

```python
starting_xis = pd.concat([match_info, players], axis=1)
starting_xis.head()
```

Look at number of unique values in each column

```python
starting_xis.nunique()
```

Drop columns with only 1 or 2 unique values

```python
unique_cols = starting_xis.nunique()
drop_cols2 = list(unique_cols[unique_cols <= 2].index)
starting_xis = starting_xis.drop(drop_cols2, axis=1)
starting_xis.head()
```

Import 'la_liga_matches' CSV so that it can be merged with the 'starting_xis' dataframe

```python
matches = pd.read_csv('../data/cleaned-data/la_liga_matches.csv')
```

```python
matches
```

```python
#Drop 'season_name' column
starting_xis = starting_xis.drop(columns='season_name')
```

```python
match_info = pd.merge(matches, starting_xis, on='match_id')
```

```python
match_info
```

Add 'match_id' column to 'players' dataframe

```python
players = players.assign(match_id = starting_xis['match_id'])
players
```

Assign dummy variables to the 'players' dataframe

```python
players_dummies = pd.get_dummies(players)
```

```python
players_dummies
```

Transpose 'player_dummies' dataframe, remove prefixes, groupby player name

```python
players_dummies_grouped = players_dummies.copy()
players_dummies_grouped.columns = players_dummies_grouped.columns.str[9:]
players_dummies_grouped.columns = players_dummies_grouped.columns.str.lstrip('_')
players_dummies_grouped = players_dummies_grouped.groupby(players_dummies_grouped.columns, axis=1).sum()
players_dummies_grouped
```

Add 'match id' column name

```python
players_dummies_grouped.rename(columns={players_dummies_grouped.columns[0]: 'match_id' }, inplace = True)
```

```python
players_dummies_grouped
```

Group dataframe to show the number of games each player has played with Lionel Messi

```python
games_sum = players_dummies_grouped.sum()
games_sum
```

Drop players that have played fewer than 20 games with Messi (38 games in La Liga season)

```python
not_enough_games = list(games_sum[games_sum < 20].index)
regular_players = players_dummies_grouped.drop(not_enough_games, axis=1)
regular_players
```

```python
matches
```

Merge 'matches' & 'regular_players' dataframes

```python
match_info_dummies = pd.merge(matches, regular_players, on='match_id')
match_info_dummies
```

Create columns showing which games Barcelona won or lost

```python
match_info_dummies['barcelona_win'] = np.where((match_info_dummies['home_score'] > match_info_dummies['away_score']) & (match_info_dummies['home_team_name'] == 'Barcelona'), 1, 0)
match_info_dummies['barcelona_loss'] = np.where((match_info_dummies['home_score'] < match_info_dummies['away_score']) & (match_info_dummies['home_team_name'] == 'Barcelona'), 1, 0)
match_info_dummies.head()
```

Select only 'player' columns and 'barcelona_win' column

```python
player_win_influence = match_info_dummies[match_info_dummies.columns[-62:]].copy()
player_win_influence = player_win_influence[player_win_influence.columns[:61]]
player_win_influence
```

Create list of all player names in 'player_win_influence' dataframe

```python
regular_player_list = list(player_win_influence.columns)
regular_player_list = regular_player_list[:-1]
```

Look at correlation between all variables

```python
player_correlation = player_win_influence.corr()
player_correlation
```

Look at correlation between each indivudual player and a barcelona win

```python
ind_player_influence = player_correlation.loc['barcelona_win']
ind_player_influence.sort_values()
```

Use OLS to look at the strength of correlation between the players and the match result

```python
X = player_win_influence[regular_player_list]
X = sm.add_constant(X)
Y = player_win_influence['barcelona_win']
```

```python
model = sm.OLS(Y,X)
results = model.fit()
predictions = results.predict(X)
```

```python
results.summary()
```

An adjusted R-squared value of 0.002 shows that there is literally no correlation between the players and match result. To improve this in the future, I could look at selecting certain players from the list and seeing how the combination of players affects the adjusted R-squared value.
