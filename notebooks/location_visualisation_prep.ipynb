```python
import pandas as pd
import numpy as np
import ast
pd.options.display.max_columns = None
```

```python
shots = pd.read_csv('../data/cleaned-data/all_shots.csv')
```

```python
shots.head()
```

```python
type(shots['location'][0])
```

```python
shots['location'] = [ast.literal_eval(i) for i in shots['location']]
```

```python
type(shots['location'][0])
```

```python
location_split = shots['location'].apply(pd.Series)
location_split
```

```python
location_split = location_split.rename(columns={0: "x_loc", 1: "y_loc"})
location_split.head()
```

```python
shots = pd.concat([shots, location_split], axis=1)
shots.head()
```

```python
shots = shots.drop(columns = ['season_name', 'location', 'type_id', 'body_part_id', 'technique_id', 'outcome_id'])
shots
```

```python
#Add 'shot probability columns'
```

```python
shot_prob = pd.read_csv('../data/cleaned-data/shot_probability.csv')
shot_prob
```

```python
shot_vis = pd.merge(shots, shot_prob, how='left', on='id')
```

```python
shot_vis
```

```python
shot_vis = shot_vis.to_csv('../data/cleaned-data/shot_vis.csv', index=False)
```

```python
shots = pd.read_csv('../data/cleaned-data/shot_vis.csv')
matches = pd.read_csv('../data/cleaned-data/la_liga_matches.csv')
```

```python
shots.shape
```

```python
shot_vis_detail = pd.merge(shots, matches, how='left', on='match_id')
shot_vis_detail.head()
```

```python
shot_vis_detail['opposition'] = np.where(shot_vis_detail['home_team_name'] == 'Barcelona', shot_vis_detail['away_team_name'], shot_vis_detail['home_team_name'])
```
