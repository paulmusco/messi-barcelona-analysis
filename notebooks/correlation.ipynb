```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
```

```python
matches = pd.read_csv('../data/cleaned-data/la_liga_matches.csv', low_memory=False)
```

```python
matches = sm.add_constant(matches)
matches.head()
```

```python
len(matches)
```

```python
#matches['barcelona_win'] = np.where((matches['home_score'] > matches['away_score']) & (matches['home_team_name'] == 'Barcelona'), 1, 0)
#matches['barcelona_loss'] = np.where((matches['home_score'] < matches['away_score']) & (matches['home_team_name'] == 'Barcelona'), 1, 0)
matches['match_points'] = np.where(
    ((matches['home_score'] > matches['away_score']) & (matches['home_team_name'] == 'Barcelona')) |
    ((matches['away_score'] > matches['home_score']) & (matches['away_team_name'] == 'Barcelona')), 3,
        np.where(((matches['home_score'] < matches['away_score']) & (matches['home_team_name'] == 'Barcelona')) | 
                 ((matches['away_score'] < matches['home_score']) & (matches['away_team_name'] == 'Barcelona')), 0, 
                     1))
matches['opposition'] = np.where(matches['home_team_name'] == 'Barcelona', matches['away_team_name'], matches['home_team_name'])
matches.head()
```

```python
#drop unwanted columns
matches = matches.drop(columns=['season_id', 'home_score', 'away_score', 'home_team_id', 'home_team_name', 'away_team_id', 'away_team_name', 'stadium_id', 'referee_id'])
matches.head()
```

```python
matches.dtypes
```

```python
matches['match_date'] = pd.to_datetime(matches['match_date'])
matches['match_date_numeric'] = pd.to_numeric(matches['match_date'])
```

```python
matches.head()
```

```python
#Bin categories, concat and drop redundant columns before correlation calculation
```

```python
season_df = pd.get_dummies(matches['season_name'], drop_first=True, prefix='season')
kick_off_df = pd.get_dummies(matches['kick_off'], drop_first=True, prefix='ko')
stadium_df = pd.get_dummies(matches['stadium_name'], drop_first=True, prefix='stad')
referee_df = pd.get_dummies(matches['referee_name'], drop_first=True, prefix='ref')
opposition_df = pd.get_dummies(matches['opposition'], drop_first=True, prefix='opp')
```

```python
matches_binned = pd.concat([matches, season_df, opposition_df, kick_off_df, stadium_df, referee_df], axis=1)
matches_binned.head()
```

```python
matches_binned = matches_binned.drop(columns=['season_name', 'kick_off', 'stadium_name', 'referee_name', 'opposition'])
matches_binned.head()
```

```python
matches_binned.shape
```

# Correlation calculation

```python
correlation = matches_binned.corr()
correlation
```

```python
ind_variables = correlation.loc["match_points"]
ind_variables = ind_variables.sort_values(ascending=False)
ind_variables
```

```python
#This shows that there is very little correlation between any of the variables and how well Barcelona perform.
```

# Look at if there is correlation between the same variables and Messi's output

```python
messi_goals = pd.read_csv('../data/cleaned-data/messi_goals.csv', low_memory=False)
messi_assists = pd.read_csv('../data/cleaned-data/messi_assists.csv', low_memory=False)
```

```python
messi_goals
```

```python
messi_assists
```

```python
messi_assists = messi_assists[['match_id', 'id', 'type_name']].copy()
messi_goals = messi_goals[['match_id', 'id', 'type_name']].copy()
```

```python
messi_goal_involvements = pd.concat([messi_goals, messi_assists])
```

```python
messi_goal_involvements
```

```python
#messi_goal_involvements.groupby(['match_id', 'type_name']).count().reset_index
```

```python
messi_grouped = messi_goal_involvements.groupby('match_id').nunique().reset_index()
messi_grouped
```

```python
messi_grouped = messi_grouped.rename(columns={"id": "goal_involvements", "type_name": "goal_assist"})
```

```python
#Merge 'messi_grouped' with 'matches_binned'
```

```python
messi_binned = matches_binned.merge(messi_grouped, how='left', on='match_id')
messi_binned
```

```python
messi_correlation = messi_binned.corr()
messi_correlation
```

```python
messi_variables = messi_correlation.loc["goal_involvements"]
messi_variables = messi_variables.sort_values(ascending=False)
messi_variables
```

```python
#Again, not much to see here in terms of correlation with individual variables
```

# Messi's performance at specific stadiums and against specific teams

```python
matches
```

```python
messi_goal_involvements
```

```python
match_goal_assist_split = messi_goal_involvements.groupby(['match_id', 'type_name']).count().reset_index()
match_goal_assist_split
```

```python
match_goal_assist_split['goals'] = np.where((match_goal_assist_split['type_name'] == 'Shot'), match_goal_assist_split['id'], 0)
match_goal_assist_split['assists'] = np.where((match_goal_assist_split['type_name'] == 'Pass'), match_goal_assist_split['id'], 0)
match_goal_assist_split = match_goal_assist_split.drop(columns=['id', 'type_name'])
match_goal_assist_split
```

```python
match_goal_assist_split['match_id'].nunique()
```

```python
involvements = match_goal_assist_split.groupby('match_id').sum()
involvements
```

```python
messi_involvement_matches = matches.merge(involvements, how='left', on='match_id')
messi_involvement_matches
```

```python
messi_involvement_matches.isnull().sum()
```

```python
#So there are 149 matches where Messi didn't score or assist a goal, and 336 matches where he registered at least one goal involvement.
```

```python
messi_involvement_matches = messi_involvement_matches.fillna(0)
```

```python
messi_involvement_matches.isnull().sum()
```

```python
messi_involvement_matches = messi_involvement_matches.sort_values(by=['match_date']).reset_index(drop=True)
messi_involvement_matches
```
