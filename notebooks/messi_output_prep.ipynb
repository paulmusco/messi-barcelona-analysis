# Goals and assists

Subset dataframe to look at goals/assists from Messi and also the players Messi assists/receives the assists from...

```python
shots = events[events['shot'].notnull()]
```

```python
shots.head()
```

```python
import ast
```

```python
shots['shot'] = [ast.literal_eval(i) for i in shots['shot']]
```

```python
all_goals = shots.loc[shots['shot'].apply(lambda x: x['outcome']['id']) == 97]
all_goals.head()
```

```python
all_goals['formation'].unique()
```

```python
all_goals = pd.concat([all_goals, shot_split], axis=1).drop(columns = ['shot'])
```
